```python
//Working with Haocheng Wang

#include <iostream>
using namespace std;
#include <vector>
#include <string>

struct Node{
    int length;
    int up, left, upleft;
};

void lcs(Node** map, string &one, string &two){
    for(int i = 0; i < two.size(); i++){
        for(int j = 0; j < one.size(); j++){
            if(one[j] == two[i]){
                map[i+1][j+1].length = map[i][j].length + 1;
                map[i+1][j+1].up = map[i+1][j+1].left = false;
                map[i+1][j+1].upleft = true;
            }
            else{
                if(map[i][j+1].length == map[i+1][j].length){
                    map[i+1][j+1].length = map[i][j+1].length;
                    map[i+1][j+1].up = map[i+1][j+1].left = true;
                    map[i+1][j+1].upleft = false;
                }
                else if(map[i][j+1].length > map[i+1][j].length){
                    map[i+1][j+1].length = map[i][j+1].length;
                    map[i+1][j+1].up = true;
                    map[i+1][j+1].upleft = map[i+1][j+1].left = false;
                }
                else{
                    map[i+1][j+1].length = map[i+1][j].length;
                    map[i+1][j+1].left = true;
                    map[i+1][j+1].upleft = map[i+1][j+1].up = false;
                }
            }
        }
    }
}

void helptraceback(Node** map, vector<string> &results, string &newstr, int row, int col, string &two){
    if(map[row][col].length == 0){
        results.push_back(newstr);
    }
    if(map[row][col].upleft == true){
        newstr.push_back(two[row-1]);
        helptraceback(map, results, newstr, row-1, col-1, two);
    }
    if(map[row][col].up == true){
        map[row][col].up = false;
        helptraceback(map, results, newstr, row-1, col, two);
    }
    if(map[row][col].left == true){
        map[row][col].left = false;
        helptraceback(map, results, newstr, row, col-1, two);
    }
    if(newstr.size() > 0 && newstr[newstr.size()-1] == two[row]){
        newstr.erase(newstr.end()-1);
    }
}

void printout(string newstr){
    for(int i = newstr.size()-1; i >= 0; i--){
        cout << newstr[i];
    }
    cout << endl;
}

void traceback(Node** map, int row, int col, string &two){
    string result;
    vector<string> results;
    helptraceback(map, results, result, row-1, col-1, two);
    for(int i = 0; i < results.size(); i++){
        bool print = true;
        for(int j = i - 1; j >= 0; j--){
            if(results[i] == results[j]){
                print = false;
            }
        }
        if(print) printout(results[i]);
    }
}

void printmap(Node** map, int row, int col){
    for(int i = 0; i < row; i++){
        for(int j = 0; j < col; j++){
            cout << map[i][j].length << " ";
        }
        cout << endl;
    }
}

int main(){
    string one = "anagram";
    string two = "agnar";
    int row = two.size()+1;
    int col = one.size()+1;
    Node** map = new Node* [row];
    for(int i = 0; i < row; i++){
        map[i] = new Node [col];
        map[i][0].length = 0;
        map[i][0].up = map[i][0].left = map[i][0].upleft = false;
    }
    for(int i = 0;  i < col; i++){
        map[0][i].length = 0;
        map[0][i].up = map[0][i].left = map[0][i].upleft = false;
    }
    lcs(map, one, two);
    //printmap(map, row, col);
    traceback(map, row, col, two);
}
```
